```python
import pandas as pd
import numpy as np

import matplotlib as plt
import seaborn as sns
sns.set_style('whitegrid')

import pandas_datareader as ppd
import pandas_datareader.data as web
from pandas_datareader import data as pdr

from datetime import datetime
```

```python
techList = ['AAPL','MSFT','AMZN','IBM']
end = datetime(2019,4,4,0,0)
start = datetime(end.year-1, end.month, end.day)
```

```python
symbols = {stock:ppd.DataReader(stock,'iex',start,end) for stock in techList}
```

```python
symbols['AAPL'].tail()
```

```python
symbols['AAPL'][['close','open']].plot(legend=True)
```

```python
maDay = [10,20,50]
sym = 'AAPL'
for ma in maDay:
    symbols['AAPL'][f'MA for {ma}'] = symbols['AAPL']['close'].rolling(window=ma).mean()
```

```python
symbols['AAPL'].head(51)
```

```python
symbols['AAPL'][['close','MA for 10','MA for 20','MA for 50']].plot(legend=True)
```

```python
symbols['AAPL']['DayReturn'] = symbols['AAPL']['close'].pct_change()
symbols['AAPL']['DayReturn'].plot(
                                  linestyle='--',
                                  marker='.'
                                 )
```

```python
sns.distplot(symbols['AAPL']['DayReturn'].dropna(),
             bins=50,
             color='g'
            )
```

```python
clsngDF = ppd.DataReader(techList,'iex',start,end)['close']
clsngDF.head()
```

```python
techReturn = clsngDF.pct_change()
techReturn[techList].plot(marker='.',
                          linestyle='--',
                          figsize=(20,4)
                         )
```

```python
sns.jointplot('AAPL',
              'AMZN',
              techReturn.dropna(),
              color='seagreen',
              marker='.'
             )
```

```python
return_fig = sns.PairGrid(techReturn.dropna())
return_fig.map_upper(sns.kdeplot,cmap='cool_d')
return_fig.map_lower(plt.pyplot.scatter,color='orange')
return_fig.map_diag(plt.pyplot.hist,bins=50)
```

```python
rets = techReturn.dropna()
plt.pyplot.scatter(rets.mean(),rets.std())
plt.pyplot.xlim([-0.001,0.002])
plt.pyplot.ylim([0.01,0.025])
plt.pyplot.xlabel('Expected Return')
plt.pyplot.ylabel('Risk')
for lable,x,y in zip(rets.columns,rets.mean(),rets.std()):
    plt.pyplot.annotate(
        lable,
        xy=(x,y),
        xytext=(50,50),
        textcoords='offset points',
        ha='right',
        va='bottom',
        arrowprops=dict(arrowstyle='-',color='red',connectionstyle='arc3,rad=-0.3')
    )
```
