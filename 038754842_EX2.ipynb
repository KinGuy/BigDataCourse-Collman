```python
cols = data.keys()
for col in cols:
    X = data.drop(col,axis=1)
    y = data[col]
    X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                        train_size = 0.75, 
                                                        random_state = 0
                                                       )
    tree = DecisionTreeRegressor(random_state=0)
    tree.fit(X_train,y_train)
    print(f'{col} score is: {tree.score(X_test,y_test)}')
```

### Outliers

#### show outliers for all features

```python
all_outliers = np.array([],dtype='int64')
for feature in log_data.keys():
    Q1 = np.percentile(log_data[feature],25)
    Q3 = np.percentile(log_data[feature],75)
    step = (Q3-Q1)*1.5
    print ("Data points considered outliers for the feature '{}':".format(feature))
    outlier_pts = log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))]
    display(outlier_pts)
    all_outliers = np.append(all_outliers,outlier_pts.index.values.astype('int64'))
```

#### get index of duplicate values and remove them

```python
unique_outliers ,indices= np.unique(all_outliers,return_inverse=True)
counts = np.bincount(indices)
outliers = unique_outliers[counts>1]
outliers
```

```python
goodData = log_data.drop(log_data.index[outliers]).reset_index(drop=True)
goodData
```

## === Modeling ===

### GMM

```python
from sklearn.mixture import GaussianMixture as GMM
from sklearn.metrics import silhouette_score
```

```python
for comp in range(2,6):
    clusterer = GMM(n_components=comp).fit(goodData)
    preds = clusterer.predict(goodData)
    score = silhouette_score(goodData,preds)
    print(f'components: {comp},  \tscore: {score}')
```

We can see thet the best (highest) score is for 2 components.
