# === Import ===

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
boston = load_boston()
sns.set_style('whitegrid')
```

```python
print(boston.DESCR)
```

# === Exploration ===

```python
data = DataFrame(boston.data,columns=boston.feature_names)
data['PRICE'] = boston.target
data.describe()
```

```python
plt.hist(data.PRICE,bins=50)
plt.xlabel('Price (K%)')
plt.ylabel('# of houses')
plt.plot()
```

```python
%matplotlib inline
plt.scatter(data['RM'],boston.target)
plt.show()
```

```python
plt.matshow(data.corr())
plt.show()
```

```python
for i in data.columns:
    data.plot(x=i,y='PRICE',kind='scatter')
```

```python
sns.lmplot('RM','PRICE',data=data)
```

```python
sns.pairplot(data, diag_kind='kde', markers='+')
```

```python
m1 = (data.corr()**2)**0.5<0.5                 #cut out all values under 0.5
m2 = np.zeros_like(data.corr(), dtype=np.bool) #cut out one diagonal
m2[np.triu_indices_from(m2)] = True            #cut out one diagonal
mask = m1+m2                                   #join two masks
sns.heatmap(data.corr(), annot=True, ax=plt.subplots(figsize=(10, 10))[1], mask=mask)
```

# === Fit Model ===

```python
# transform data to vectors
X = np.vstack(data.RM)
Y = data.PRICE
#X = np.array([[v,1] for v in X])
#print(X)
print(np.linalg.lstsq(np.array([[x,1] for x in data.RM]),Y))
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lreg.fit(data.drop('PRICE',1),data.PRICE)
```

```python
print(lreg.coef_) # bigger = better
```

```python
lreg.intercept_
```

```python
lreg.score(data.drop('PRICE',1),data.PRICE) #R^2 score
```

```python
sns.scatterplot(lreg.predict(data.drop('PRICE',1)),data.PRICE)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.drop('PRICE',1), data.PRICE, test_size=0.3)
```

```python
lreg = LinearRegression()
lreg.fit(x_train,y_train)
print(lreg.coef_)
print(lreg.score(x_test,y_test))
sns.scatterplot(lreg.predict(x_test),y_test)
```

```python
lreg = LinearRegression()
lreg.fit(data.drop('PRICE',1),data.PRICE)
print(lreg.coef_)
print(lreg.score(data.drop('PRICE',1),data.PRICE))
sns.scatterplot(lreg.predict(data.drop('PRICE',1)),data.PRICE)
```

```python
poly = PolynomialFeatures(degree=2,include_bias=True)
data3 = pd.DataFrame(poly.fit_transform(data.drop('PRICE',1)))

lreg = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(data3, data.PRICE, test_size=0.3)
lreg.fit(x_train,y_train)
print(lreg.coef_)
print(lreg.score(x_test,y_test))
sns.scatterplot(lreg.predict(x_test),y_test)
```
